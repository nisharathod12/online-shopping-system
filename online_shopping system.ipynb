```python
### import string
import datetime

class shopping:
    def __init__(self):
        print('===============================================')
        print('          ! ! WELCOME TO OUR SHOP ! !          ')
        print('===============================================')
        self.menu1()
    def menu1(self):
        choice=input("""HOW CAN I HELP YOU ?
        Press 1. For Go To Admin Page...
        Press 2. For Go To User Login Page...
        press 3. for Exit..
        Enter Your Choice...:""")
        if(choice=='1'):
            adm=Admin()
        if (choice=='2'):
            log=login()
        if(choice=='3'):
            print('=====================================')
            print('     ! ! THANK YOU FOR VISIT ! !     ')
            print('=====================================')
            exit()

class login(shopping):
    def __init__(self):
        self.user_name=" "
        self.password=" "
        self.total_bill=0
        self.menu()
        self.menu1()
    
    def menu(self):
        choice=int(input("""HOW CAN I HELP YOU ?
        press 1. for create a account..
        press 2. for see user details..
        press 3. for add product in user cart..
        press 4. for update product quantity..
        press 5. for delete product in user cart..
        press 6. for see all product details..
        press 7. for generate bill..
        press 8. for pay bill..
        Enter your choice."""))
        if choice==1:
            if not self.user_name.strip():
                self.create_account()
            else:
                print("Account already exists!")
        if choice==2:
            self.user_details()
        if choice==3:
            self.add_product_c()
        if choice==4:
            self.update_product_c()
        if choice==5:
            self.delete_product_c()
        if choice==6:
            self.cart_details()
        if choice==7:
            self.generate_bill()
        if choice==8:
            self.pay_bill()
        
    def create_account(self):
        name=input('Enter Your user name..:')
        password=input('Enter Your password..:')
        a=0
        d=0
        if len(password)>=8:
            for i in password:
                if i.isdigit():
                    d+=1
                elif i.isalpha():
                    a+=1
                else:
                    pass
        else:
            print(' your password length is must be 8')
            self.menu()
        if   a>=1  and d>=1:
            print("your account created successfully..:")
        else:
            print(" your password must contain digit and alphabet ")
            
        
        self.user_name=name
        self.password=password
        self.menu()

    def user_details(self):
        print('User Name :',self.user_name)
        print('password :',self.password)
        self.menu()

    def add_product_c(self):
        product_name = input('Enter the product name to add to your cart: ')
        selected_product = next((product for product in Admin.products if product['product_name'] == product_name), None)

        if selected_product:
            quantity = int(input('Enter the quantity to add to your cart: '))
            if 'cart' not in dir(self):
                self.cart = {}

            
            if selected_product['product_stock'] >= quantity:
                if product_name in self.cart:
                    self.cart[product_name] += quantity
                else:
                    self.cart[product_name] = quantity

                selected_product['product_stock'] -= quantity
                self.update_stock_file()  
                print(f'{quantity} {selected_product["product_name"]} added to your cart.')
            else:
                print(f'Insufficient stock for {selected_product["product_name"]}. Please choose a smaller quantity.')
        else:
            print('Product not found.')

        self.menu()

    def update_stock_file(self):
        with open("stock_product.txt", "w") as file:
            for product in Admin.products:
                file.write(str(product) + "\n")

    def update_product_c(self):
        product_name = input('Enter the product name to update quantity in your cart: ')
        if 'cart' in dir(self) and product_name in self.cart:
            new_quantity = int(input('Enter the new quantity: '))
            selected_product = next((product for product in Admin.products if product['product_name'] == product_name), None)
            if selected_product:
                if selected_product['product_stock'] >= new_quantity:
                    self.cart[product_name] = new_quantity
                    print(f'Quantity updated for product {product_name}.')
                else:
                    print(f'Insufficient stock for {product_name}. Please choose a smaller quantity.')
            else:
                print('Product not found.')
        else:
            print('Product not found in your cart.')

        self.menu()
    def delete_product_c(self):
        product_name = input('Enter the product name to delete from your cart: ')
        if 'cart' in dir(self) and product_name in self.cart:
            del self.cart[product_name]
            selected_product = next((product for product in Admin.products if product['product_name'] == product_name), None)
            if selected_product:
                print(f'Product {product_name} deleted from your cart.')
                selected_product['product_stock'] += self.cart[product_name]  
            else:
                print('Product not found .')
        else:
            print('Product not found in your cart.')

        self.menu()


    def cart_details(self):
        if 'cart' in dir(self) and self.cart:
            print('--------- Your Shopping Cart ---------')
            for product_name, quantity in self.cart.items():
                selected_product = next((product for product in Admin.products if product['product_name'] == product_name), None)
                if selected_product:
                    product_name = selected_product['product_name']
                    product_price = selected_product['product_price']
                    product_total = product_price * quantity
                    print(f'{product_name} (Quantity: {quantity}) - Total: ₹{product_total}')
            print('-------------------------------------')
        else:
            print('Your cart is empty.')

        self.menu()

    def generate_bill(self):
        if 'cart' in dir(self) and self.cart:
            current_time = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            print('--------- Your Shopping Cart ---------')
            for product_name, quantity in self.cart.items():
                selected_product = next((product for product in Admin.products if product['product_name'] == product_name), None)
                if selected_product:
                    product_name = selected_product['product_name']
                    product_price = selected_product['product_price']
                    product_total = product_price * quantity
                    self.total_bill += product_total
                    print(f'{product_name} (Quantity: {quantity}) --- Total: ₹{product_total}')

            print(f'-------------------------------------')
            print(f'Total Bill: ₹{self.total_bill}')
            print('-------------------------------------')

            
            self.generate_bill_text(current_time, self.total_bill)
            total_bills=self.total_bill

        else:
            print('Your cart is empty.')

        self.menu()

    def generate_bill_text(self, current_time, total_bills):
        with open("bill.txt", "w") as file:
            file.write(f"Bill generated on: {current_time}\n")
            file.write('=========================================bill=====================================   \n')
            file.write('----------------------------------------------------------------------------------    \n')
            for product_name, quantity in self.cart.items():
                selected_product = next((product for product in Admin.products if product['product_name'] == product_name), None)
                if selected_product:
                    product_name = selected_product['product_name']
                    product_price = selected_product['product_price']
                    product_total = product_price * quantity
                    file.write(f'Name:{product_name}        Quantity:{quantity}       Price:{product_price}      TOtal:{product_total}\n')

            file.write('------------------------------------------------------------------------------------    \n')
            file.write(f'                                 Total Bill: {self.total_bill}                               \n')
            file.write('====================================================================================     \n')

        print('Bill generated successfully. Check "bill.txt" for details.')

    def pay_bill(self):
        given_cash=int(input('Enter money you have to pay'))
        if self.total_bill>given_cash:
            print('Insufficient funds.')
            give_amount=self.total_bill-given_cash
            print(' ₹',give_amount,'Money you want to pay')
            self.menu()
        else:
            change_amount=given_cash-self.total_bill
            print(' ₹',change_amount,'Money you get back')
        
        

class Admin(shopping):
    products = []

    def __init__(self):
        self.admin_name = 'admin'
        self.admin_password = '1234'
        self.load_products() 
        self.menu()
        self.menu1()

    def menu(self):
        user_input = int(input("""HOW CAN I HELP YOU ?
        Press 1. for change credentials of admin..
        press 2. for add new product..
        press 3. for update stock of particular product..
        press 4. for see all product details..
        press 5. for delete product..
        press 6. for search product by name..
        press 7. for exit..
        Enter Your Choice..:"""))
        if user_input == 1:
            self.change_credential()
        elif user_input == 2:
            self.add_product()
        elif user_input == 3:
            self.update_product()
        elif user_input == 4:
            self.product_details()
        elif user_input == 5:
            self.delete_product()
        elif user_input == 6:
            self.search_product()
        elif user_input == 7:
            exit()
        else:
            print("Invalid choice. Please try again.")
            self.menu()

    def change_credential(self):
        a_name = input("enter admin name..:")
        a_password = input('enter admin password..:')

        if self.admin_name == a_name and self.admin_password == a_password:
            print('Your admin data is correct')
            new_a_name = input('enter admin new name..:')
            new_a_password = input('enter admin new password..:')
            self.admin_name = new_a_name
            self.admin_password = new_a_password
            print("admin name:", self.admin_name)
            print("admin password:", self.admin_password)
            self.menu()
        else:
            print('Your admin data is incorrect')
            self.menu()

    def add_product(self):
        name = input('enter admin name: ')
        password = input('enter admin password: ')
        if self.admin_name == name and self.admin_password == password:
            product_name = input('enter name of product: ')
            product_category = input('enter category of the product: ')
            product_stock = int(input('enter available stock of the product: '))
            product_price = int(input('enter price of the product: '))

            product = {
                'product_name': product_name,
                'product_category': product_category,
                'product_stock': product_stock,
                'product_price': product_price
            }
            Admin.products.append(product)
            self.save_products()
            print('Product added successfully!')
            self.menu()
        else:
            print('Your admin data is incorrect')
            self.menu()

    def update_product(self):
        p_name = input('Enter product name you want to update stock: ')
        for product in Admin.products:
            if product['product_name'] == p_name:
                update_s = int(input("Enter update stock: "))
                product['product_stock'] += update_s
                self.update_stock_file()  
                print('Product stock updated successfully!')
                self.menu()
                return
        print("Enter valid product name")
        self.menu()

    def update_stock_file(self):
        with open("stock_product.txt", "w") as file:
            for product in Admin.products:
                file.write(str(product) + "\n")

    def product_details(self):
        for product in Admin.products:
            print('product name:', product['product_name'])
            print('product category:', product['product_category'])
            print('product stock:', product['product_stock'])     
            print('product price:', product['product_price'])
            print('----------------------------')
        self.menu()

    def delete_product(self):
        product_to_delete = input('Enter product name you want to delete: ')
        for product in Admin.products:
            if product['product_name'] == product_to_delete:
                Admin.products.remove(product)
                self.save_products()
                print('Product deleted successfully!')
                self.menu()
                return
        print('Cannot delete, invalid product name.')
        self.menu()

    def search_product(self):
        name = input('Enter product name you want to search: ')
        found = False
        for product in Admin.products:
            if product['product_name'].lower() == name.lower():
                print('Product Name:', product['product_name'])
                print('Product Category:', product['product_category'])
                print('Product Stock:', product['product_stock'])
                print('Product Price:', product['product_price'])
                print('----------------------------')
                self.menu()
                found = True
        if not found:
            print('Product not found.')
            self.menu()

    def load_products(self):
        try:
            with open("stock_product.txt", "r") as file:
                data = file.readlines()
                for line in data:
                    product = eval(line.strip())
                    Admin.products.append(product)
        except FileNotFoundError:
            print("Product file not found. Starting with an empty product list.")
            
    def save_products(self):
        with open("stock_product.txt", "w") as file:
            for product in self.products:
                file.write(str(product) + "\n")

shp=shopping()
```
